```python
from matplotlib import pyplot as plt
import gdal
from sklearn import cluster
```

```python
img = gdal.Open("E:\GIS\GRACE\Grace\GRCTellus.JPL.20161225.LND.RL05_1.DSTvSCS1411.tif\GRCTellus.JPL.20161225.LND.RL05_1.DSTvSCS1411.tif")
data = img.ReadAsArray()
X = data.reshape((-1,1))
```

```python
k_means = cluster.KMeans(n_clusters=8)
k_means.fit(X)
X_cluster = k_means.labels_
X_cluster = X_cluster.reshape(data.shape)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.imshow(data)
ax2.imshow(X_cluster, cmap="hsv")
```
